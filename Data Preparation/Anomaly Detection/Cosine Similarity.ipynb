# Importing Libraries

```python
# Imports Libraries
import numpy as np
import matplotlib.pyplot as plt
```

# Initalising Data

```python
# For linear data
xData = np.array([1, 2, 2, 3, 4, 5, 6, 7, 13])
yData = np.array([1, 2, 3, 3, 5, 5, 7, 7, 8])

# Calculates gradients relative to a point
data = yData / xData
```

# Plotting Box Plot

```python
plt.figure(figsize=(10, 1.2))

plt.xlabel('X LABEL')
plt.yticks([0])

plt.boxplot(data, vert=0, widths=0.5)

plt.show()
```
