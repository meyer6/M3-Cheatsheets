# Importing Libraries

```python
# Imports Libraries
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Initialising Data

```python
featureNames = ['Class Name 1', 'Class Name 2', 'Class Name 3']

xData = np.array([[-1, -1, 2], [-2, -1, 4], [-3, -2, 1], [1, 1, 3], [2, 1, -2], [3, 2, -4]])

# Scales the data
xData = StandardScaler().fit_transform(xData)
```

# Model

```python
# Creates the model
pca = PCA()
pca.fit(xData)
xPCA = pca.transform(xData)
```

# Plotting the Results

```python
# Gets the ratios of the explained variance
ratios = pca.explained_variance_ratio_[::-1]

# Initialises settings of the graph
cMap = plt.get_cmap('Blues')

plt.figure(figsize=(7,2))

# Plots the ratios of the explained variance
plt.barh(range(len(ratios)), ratios,
         color=cMap(ratios * 3 / 4 + 0.25))
plt.yticks(range(len(ratios)), [f'PC{i}' for i in range(len(ratios), 0, -1)], fontsize=12)

plt.xticks(fontsize=12)
plt.xlabel('Proportion of Explained Variance', fontsize=12)

plt.show()
```

```python
# Initialises settings of the graph
plt.figure(figsize=(7,4))

plt.xlabel('PC1')
plt.ylabel('PC2')

# Plots the PC1 and PC2
plt.scatter(xPCA[:,0], xPCA[:,1], marker="o", s=25)

plt.show()
```
