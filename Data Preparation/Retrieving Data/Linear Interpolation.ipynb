```python
import numpy as np
```

```python
data = np.array([98, 173, 279, 422, 588, 716, np.NaN, 907, 1139, 1364, 1637, 2074, 2767, 3397])

# Retrieves indexes of NaN data points
notNan = np.logical_not(np.isnan(data))
indices = np.arange(len(data))

# Interpolates over NaN data points
data = np.interp(indices, indices[notNan], data[notNan])
```
