# Initialising Data

```python
# The different states that are possible
states = ["Class 1", "Class 2", "Class 3"]

# Defines the inital distribution
currentState = np.array([10, 3, 4])

# Defines the transition matrix
transitionMatrix = np.array([[0.3, 0.2, 0.5],
                             [0.6, 0.3, 0.1],
                             [0.2, 0.4, 0.4]])

# Transposes the transition matrix for later multiplication
transposedTransitionMatrix = np.transpose(transitionMatrix)

ITERATIONS = 10
```

```python
# Calculates the new states over a number of iterations
allStates = [currentState]
for _ in range(ITERATIONS):
    currentState = transposedTransitionMatrix.dot(currentState)
    allStates.append(currentState)
```

# Plotting Data

```python
X_LABEL = "x-axis label"
Y_LABEL = "y-axis label"

# Initialises settings of the graph
plt.figure(figsize=(12,3))

plt.xlabel(X_LABEL)
plt.ylabel(Y_LABEL)

# For each state, plots the points after each iteration
for i, state in enumerate(np.array(allStates).T):
  plt.plot(np.linspace(0, len(state), len(state)), state, marker=".", label=states[i])

plt.legend()
plt.show()
```
