```python
# Imports libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

# Initaliasing Data

```python
# Gives example data
from sklearn.datasets import make_regression
data = make_regression(n_features=4, n_samples=1000, n_informative=4, noise=2)

featureNames = np.array(["Class Name 1", "Class Name 2", "Class Name 3", "Class Name 4"])
outputName = "Class Name 5"
xData, yData = data
data = np.concatenate((xData, yData.reshape(-1, 1)), axis=1)

df = pd.DataFrame(data=data,
            columns=np.append(featureNames, np.array([outputName])))

# Calculates the correlation matrix
corrMatrix = df.corr().abs()
```

# Model

```python
plt.figure(figsize=(4,3), dpi=200)

# Plots the correlation matrix
ax = sb.heatmap(corrMatrix, annot=True, cmap="Greens")
ax.tick_params(axis='x', rotation=45)
```
