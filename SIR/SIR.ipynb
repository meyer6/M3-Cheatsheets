# Importing Libraries

```python
# Imports libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Initialising Data

```python
# Initialises variables
a = 0.2
b = 0.1
c = 0.05

S0 = 1
I0 = 50 / 109_581_078
R0 = 0

y0 = S0, I0, R0

START_YEAR, END_YEAR = 2020, 2100
```

# Model

```python
# Sets up differential equations
def differentials(y, t):
    S, I, R = y
    dSdt = -a * S * I
    dIdt = a * S * I - b * I
    dRdt = b * I
    return dSdt, dIdt, dRdt

# Solves differential equations and
# calculates range of S, I and R values in time interval
t = np.linspace(0, END_YEAR - START_YEAR, 500)
S, I, R = odeint(differentials, y0, t).T

t += START_YEAR
```

# Plotting Data

```python
X_LABEL = "x-axis label"
Y_LABEL = "y-axis label"

# Initialises settings of the graph
plt.figure(figsize=(12,3))

# Plots the lines
plt.plot(t, S + I + R, label='Total Population')
plt.plot(t, S, label='S(t)')
plt.plot(t, I, label='I(t)')
plt.plot(t, R, label='R(t)')

plt.xlabel(X_LABEL)
plt.ylabel(Y_LABEL)

plt.legend()
plt.show()
```
